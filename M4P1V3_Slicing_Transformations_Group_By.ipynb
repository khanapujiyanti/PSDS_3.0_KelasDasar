#GroupBy Satu Variabel dengan Satu Fungsi Aggregat

Kita dapat melakukan groupbyterhadap satu buah variabel (contoh: category) dengan menggunakan sebuah fungsi aggregat (mean).

````python
#GroupBy berdasarkan variabel 'category', fungsi aggregat 'mean'
hasil_group_by=df.groupby('category').mean()
hasil_group_by
````

````python
#GroupBy erdasarkan variabel 'category', fungsi aggregat 'mean'
hasil_group_by= df.groupby('category')[['sales', 'profit']].mean()
hasil_group_by
````

#GRoupBy Multivariabel dengan Satu Fungsi Aggregat

````python
sub_category=df.groupby(['category','sub_category'])[['sales','quantity','profit']].mean()
sub_category
````

#GroupBy dengan Banyak Fungsi Aggregat

Untuk menampung banyak fungsi aggregat, atau ketika kita menginginkan fungsi aggregat yang berbeda-beda pada masing-masing variabel, kita dapat menuliskannya dalam sebuah dictionary sebagai berikut:

````python
import numpy as np
````

````python
sub_category=df.groupby(['category'])[['sales','quantity','profit']].agg({
    'sales':min,
    'quantity':max,
    'profit': np.mean
})
sub_category
````

#GroupBy dengan Fungsi Aggregat Custom

````python
def data_range(series):
  return series.max() - series.mean()
sub_category=df.groupby(['sub_category','category'])[['sales', 'quantity', 'profit']].agg(data_range)
sub_category
````
